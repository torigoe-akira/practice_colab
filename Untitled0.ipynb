```python
from re import X
class DataManager:
  def __init__(self,x,y,z):
    self.x=x
    self.y=y
    self.z=z
  def add_x(self,delta):
    self.x += delta
  def add_y(self,delta):
    self.y += delta
  def add_z(self,delta):
    self.z += delta
  def sum(self):
    return sum([self.x,self.y,self.z])
data_manager = DataManager(2,3,5)
print(data_manager.sum())
data_manager.add_x(4)
print(data_manager.sum())
data_manager.add_y(0)
print(data_manager.sum())
data_manager.add_z(-9)
print(data_manager.sum())
```

```python
x=[[1,2,3],[4,5,6]]
y=[[8,1,2],[-1,0,-2]]
x_y_1=[]
def matrix_sum(x,y):
  for int_x1,int_y1 in zip(x,y):
    x_y_2=[]
    #print(int_x1)
    for int_x2,int_y2 in zip(int_x1,int_y1):
      x_y_2.append(int_x2+int_y2)
    x_y_1.append(x_y_2)
  return x_y_1
answer=matrix_sum(x,y)
print(answer)
```

```python
xy=[]
def matrix_vector_product(x,y):
  for int_x_1 in x:
    xy.append(sum(int_x_2*int_y for int_x_2,int_y in zip(int_x_1,y)))
  return xy
x=[[1,2,3],[4,5,6]]
y=[8,1,2]
print(matrix_vector_product(x,y))
```

```python
x=[[1,2,3],[4,5,6]]
y=[[8,1],[-1,0],[0,1]]
xy=[]
y=list(zip(*y))
def matrix_product(x,y):
  for int_x_1 in x:
    Nrow=[]
    for int_y_1 in y:
        Nrow.append(sum(int_y_2*int_x_2 for int_x_2,int_y_2 in zip(int_x_1,int_y_1)))
    xy.append(Nrow)
  return xy
print(matrix_product(x,y))
```
